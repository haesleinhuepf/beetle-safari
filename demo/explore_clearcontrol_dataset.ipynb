```python
# Initialize GPU
import pyclesperanto_prototype as cle
import beetlesafari as bs
import numpy as np
from skimage.io import imread

cle.select_device('RTX')
```

```python
cc_dataset = bs.ClearControlDataset('C:/structure/data/2019-12-17-16-54-37-81-Lund_Tribolium_nGFP_TMR/')

print("Duration in seconds:", cc_dataset.get_duration_in_seconds())

print("Duration in hours:", bs.seconds_to_hours(cc_dataset.get_duration_in_seconds()))
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 8, figsize=(15, 7), sharey=True)

# visualize the dataset at the beginning of every day
for d in range(0, 8):
    image = cc_dataset.get_resampled_image(time_in_seconds=bs.days_to_seconds(d))
    cle.imshow(image, plot=axs[d])

fig.suptitle('7 days of Tribolium embryo development')
```

```python
# pick a single stack and show projections of the 3D data
image = cc_dataset.get_resampled_image(time_in_seconds=bs.hours_to_seconds(20))

fig, axs = plt.subplots(1, 3, figsize=(15, 7))

# show projections
cle.imshow(cle.maximum_x_projection(image), plot=axs[0])
cle.imshow(cle.maximum_y_projection(image), plot=axs[1])
cle.imshow(cle.maximum_z_projection(image), plot=axs[2])
```
