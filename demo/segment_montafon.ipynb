```python
from skimage.io import imread, imsave
import pyclesperanto_prototype as cle
import numpy as np
import pandas as pd
from napari_pyclesperanto_assistant._advanced_statistics import advanced_statistics

cle.select_device("TX")
```

```python
dataset = "montafon"
```

```python
image = imread(f"C:/structure/data/{dataset}.tif")
image.shape
```

```python
result_image = []
result_labels = []
table = None
for t in range(0, image.shape[0], 2):
    print("t = ", t)
    image0_s = cle.scale(image[t], None, 0.69, 0.69, 3.0, False, False, True)
    result_image.append(np.asarray(image0_s))
    print("a")
    
    image1_thb = cle.top_hat_box(image0_s, None, 5.0, 5.0, 5.0)
    print("b")
    image2_vol = cle.voronoi_otsu_labeling(image1_thb, None, 1.0, 1.0)
    print("c")
    image3_esl = cle.exclude_small_labels(image2_vol, None, 20.0)
    print("d")
    
    cle.imshow(image3_esl, labels=True)
    result_labels.append(np.asarray(image3_esl))
    
    print("e")

    #new_table = advanced_statistics(image0_s, image3_dl)
    #new_table['frame'] = t
    #if table is None:
    #    table = pd.DataFrame(new_table)
    #else:
    #    table = pd.concat([table, pd.DataFrame(new_table)], ignore_index=True)
    #print("f")

        
imsave(f"C:/structure/data/{dataset}_scaled.tif", np.asarray(result_image).astype(np.uint16))
imsave(f"C:/structure/data/{dataset}_labels.tif", np.asarray(result_labels).astype(np.uint16))
```
