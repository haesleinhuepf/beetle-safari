```python
from skimage.io import imread, imsave
import pyclesperanto_prototype as cle
import numpy as np
import pandas as pd
from napari_pyclesperanto_assistant._advanced_statistics import advanced_statistics

cle.select_device("gfx")
```

```python
image = imread("C:/structure/data/heide.tif")
image.shape
```

```python
result_image = []
result_labels = []
table = None
for t in range(0, image.shape[0], 1):
    print("t = ", t)
    image0_s = cle.scale(image[t], None, 0.69, 0.69, 3.0, False, False, True)
    result_image.append(np.asarray(image0_s))
    print("a")
    
    image1_thb = cle.top_hat_box(image0_s, None, 3.0, 3.0, 3.0)
    print("b")
    image2_vol = cle.voronoi_otsu_labeling(image1_thb, None, 0.5, 1.0)
    print("c")
    image3_dl = cle.dilate_labels(image2_vol, None, 10.0)
    print("d")
    
    cle.imshow(image3_dl, labels=True)
    result_labels.append(np.asarray(image3_dl))
    
    print("e")

    #new_table = advanced_statistics(image0_s, image3_dl)
    #new_table['frame'] = t
    #if table is None:
    #    table = pd.DataFrame(new_table)
    #else:
    #    table = pd.concat([table, pd.DataFrame(new_table)], ignore_index=True)
    #print("f")

        
imsave("C:/structure/data/heide_scaled.tif", np.asarray(result_image))
imsave("C:/structure/data/heide_labels.tif", np.asarray(result_labels))

table
```

```python
table.to_csv("C:/structure/data/heide_measurements.csv")
```
